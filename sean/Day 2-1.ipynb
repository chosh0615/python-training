```python
def maxmindiff(series):
    return series.max() - series.min()
```

```python
NC15[['타석','안타','홈런']].apply(maxmindiff)
```

### 연습

각 선수 타율 구하기 함수를 정의해 적용

```python
def 타율함수(series):
    return (series.안타 + series.홈런) / series.타석
```

```python
타율 = NC15.apply(타율함수, axis = 1)
```

```python
'{:.3f}'.format(0.38153)
```

```python
def 소수점자리수설정(값): 
    return '{:.3f}'.format(값)
```

```python
타율.map(소수점자리수설정)
```

```python
NC15[['타석','안타','홈런']].applymap(소수점자리수설정)
```
