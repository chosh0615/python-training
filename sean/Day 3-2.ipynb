# 엑셀파일

```python
경로 = 'data/koreaexim/international/Global+Market+Daily+Table%28%EA%B0%92%EB%B3%B5%EC%82%AC%29_0803.xls'
```

```python
프레임 = pd.read_excel(경로, skiprows=3, parse_cols='C:K', header=None)
프레임
```

```python
프레임 = 프레임.dropna(how='all')
```

```python
프레임 = 프레임.reset_index(drop=True)
프레임
```

```python
def 지표추출(프레임):
    표목록 = []
    표범위목록 = [(0,8), (8,16), (16,23)]
    for 범위 in 표범위목록:
        시작, 끝 = 범위
        표 = 프레임[시작:끝]
        표목록.append(표)
    return 표목록
```

```python
표목록 = 지표추출(프레임)
```

```python
표목록
```

```python
def 엑셀로내보내기(표목록, 파일명):
    엑셀파일 = pd.ExcelWriter(파일명)

    for i, 표 in enumerate(표목록):
        표.to_excel(엑셀파일, '지표 {}'.format(i+1))

    엑셀파일.save()
```

```python
엑셀로내보내기(표목록, '지표 1.xlsx')
```

모든 대상 파일에 현재 작업을 적용하기

```python
import os
```

```python
항목들 = os.listdir('data/koreaexim/international/')
```

```python
항목들
```

대상 파일 경로 구성

```python
대상파일목록 = []
폴더 = 'data/koreaexim/international/'
for 항목 in 항목들:
    경로 = os.path.join(폴더, 항목)
    대상파일목록.append(경로)
```

```python
대상파일목록
```

```python
import exim
```

```python
if not os.path.exists('output'):
    os.makedirs('output')

for i, 경로 in enumerate(대상파일목록):
    # 1. 데이터프레임 생성
    # 2. 표 추출
    표목록 = exim.지표추출(경로)
    # 3. 결과 엑셀 파일로 저장
    exim.엑셀로내보내기(표목록, '지표{}.xlsx'.format(i+1))
```
