Time-series. 시간의 흐름에 따라 기록된 자료

```python
import pandas as pd
```

```python
식당 = pd.read_excel('data/food_order.xlsx')
```

```python
식당.date[:5]
```

```python
식당.date[0]
```

```python
날짜 = 식당.date.map(lambda 날짜: '{0.year}년 {0.month}월 {0.day}일'.format(날짜))
```

특정 개수만큼 선택

```python
pd.date_range('2016-12-16', periods=20)
```

```python
pd.date_range(end='2016-12-16', periods=20)
```

```python
pd.date_range('2016-08-15 09:43:52', periods=10, normalize=True)
```

```python
pd.date_range('2016-01-01', '2017-01-01', freq='M')
```

```python
pd.date_range('2016-08-15 09:00', '2016-08-15 18:00', freq='1h30min')
```
