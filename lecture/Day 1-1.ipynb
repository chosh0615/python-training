열 선택

"열이 너무 많아서, 일부만 선택하려구요"

```python
열선택 = ['corner', 'dine_type', 'menu', 'use_count']
식당 = 식당[열선택]
식당[:3]
```

지난 3년간 코너별 주문수량 평균

먼저 코너별로 분류합니다.

```python
코너별 = 식당.groupby('corner')
```

각 코너별로 분류된 자료들에서, 주문수량(use_count)의 평균을 구합니다.

```python
코너별['use_count'].mean()
```

두 개 이상의 기준으로 분류하는 것도 가능합니다.

```python
코너별시간대별 = 식당.groupby(['corner', 'dine_type'])
```

코너별 시간대별 주문수량 평균

```python
코너별시간대별['use_count'].mean().unstack()
```

```python
코너별시간대별_주문평균 = 식당.pivot_table(
    'use_count', aggfunc='mean',
    index='corner', columns='dine_type')
```

분석 결과는 엑셀 파일로 내 보내놓으면 편리합니다.

```python
코너별시간대별_주문평균.to_excel('코너별 시간대별 주문평균.xlsx')
```

```python
코너별시간대별_주문평균 = pandas.read_excel(
    '코너별 시간대별 주문평균.xlsx')
```

```python
코너별시간대별_주문평균 = 코너별시간대별_주문평균.set_index('corner')
```
