각 식품의 영양소 정보를 데이터프레임으로 만들고, 모두 합치기

```python
frames = []

for rec in db:
    # 각 식품의 영양소 정보를 데이터프레임으로 생성
    fnuts = pd.DataFrame(rec['nutrients'])
    # 각 식품의 ID를 열로 추가
    fnuts['id'] = rec['id']
    frames.append(fnuts)
    
nutrients = pd.concat(frames, ignore_index=True)
```
